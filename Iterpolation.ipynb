```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Lagrange(x, y, g):
    #caso o grau tenha sido passado
    if g:
        #se o grau for diferente do tamanho do vetor de coordenadas
        if (len(x) != (g+1)): 
            ind = np.linspace(0, len(x)-1, g+1)
            ind = [int(round(i)) for i in ind]
            x_filter = x[ind]
            y_filter = y[ind]
            coef = np.polyfit(x_filter, y_filter, g)
            
        #se o grau for o mesmo valor do tamanho do vetor de coordenadas
        else:
            coef = np.polyfit(x,y,g);
    #se o grau nao tiver sido passado
    else:    
        coef = np.polyfit(x,y,len(x));
        
    plt.plot(x,np.polyval(coef,x),'-r')
    return coef
```

```python
#exercicio 2 - a
dia = [6,7,8,9,10,12,13,14,15,16]
dia = np.array(dia)
casos = [778,1200,1500,1800,977,2300,5200,3500,3600,3200]
casos = np.array(casos)
```

```python
coef5 = Lagrange(dia,casos,5)
np.polyval(coef5,11)
```

```python
#exercicio 3
def funcpol(x,op):
    if(op):
        return x**2 + np.sin(x) + np.cos(3*x)
    else: 
        return 1/(1+(25*(x**2)))


```
